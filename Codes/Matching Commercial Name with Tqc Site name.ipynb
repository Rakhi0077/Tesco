```python
commercial_only_OBL=df_Original_List[df_Original_List['OBL Count']>0]
commercial_only_OBL.head(2)
```

```python
jb_comm=df_To_beMatched['Junior_Buyer'].unique()
jb_comm=pd.DataFrame(jb_comm,columns={'Junior_Buyer'})
# jb_comm=jb_comm[jb_comm['Junior_Buyer']=='PREPVE        G.']
jb_comm
```

```python
cols=['TQC_Supplier_Name','TQC_Site_Name','Supplier_Code','Comm_Supplier_Name','junio_buyer','correct_ratio']
commercial_for_loop=commercial_only_OBL.copy()
df2=pd.DataFrame()

output_list=[]


for i in jb_comm['Junior_Buyer']:
    filter_data_tqc=df_To_beMatched[df_To_beMatched['Junior_Buyer']==i]
    filter_data_comm=commercial_for_loop[commercial_only_OBL['Junior_Buyer']==i]

    if not filter_data_comm.empty:

        str2Match = filter_data_tqc['Site_Name_2'].fillna('######').tolist()
        strSupplier_Name = filter_data_tqc['Supplier_Name_2'].fillna('######').tolist()
        strSupplier_Code =filter_data_tqc['Supplier_Code'].tolist()
        strOptions =filter_data_comm['Supplier_Name_2'].fillna('######').tolist()
#         strTesco_Brand_Ind =filter_data_tqc['Tesco_Brand_Ind'].fillna('######').tolist()
        
#         name_match,ratio_match=checker(str2Match,strOptions)
#         k=0
        for j in range(0,len(str2Match)):
            x=process.extractOne(str2Match[j],strOptions,scorer=fuzz.token_set_ratio)

            while x[1]==100:
                output_list=[strSupplier_Name[j],str2Match[j],strSupplier_Code[j],x[0],i,x[1]]
                zipped = zip(cols, output_list)
                a_dictionary = dict(zipped)
                df1=pd.DataFrame(a_dictionary,columns=cols,index=[j])
                df2=df2.append(df1)
                #drop row which has the matched supplier name and junor buyer
                strOptions.remove(x[0])
                x=process.extractOne(str2Match[j],strOptions,scorer=fuzz.token_set_ratio)
            strOptions =filter_data_comm['Supplier_Name_2'].fillna('######').tolist()
            
            output_list=[strSupplier_Name[j],str2Match[j],strSupplier_Code[j],x[0],i,x[1]]
            zipped = zip(cols, output_list)
            a_dictionary = dict(zipped)
            df1=pd.DataFrame(a_dictionary,columns=cols,index=[j])
            df2=df2.append(df1)

        
        
df2
matches_using_fuzzy=df2
matches_using_fuzzy=matches_using_fuzzy.drop_duplicates()
# matches_using_fuzzy[matches_using_fuzzy['TQC_Supplier_Name']=='SAMWORTH BROS']
```

```python
df_3=matches_using_fuzzy.merge(commercial_only_OBL[['Supplier_Name_2','Supplier_Number','Junior_Buyer']],how='left',left_on=['Comm_Supplier_Name','junio_buyer'],right_on=['Supplier_Name_2','Junior_Buyer'])
df_3=df_3.drop_duplicates()
df_3
matches_using_fuzzy_with_Supplier_Number=df_3
matches_using_fuzzy_with_Supplier_Number.head(2)
# matches_using_fuzzy_with_Supplier_Number.to_excel('df4.xlsx')
```

```python
matches_using_fuzzy_with_Supplier_Number['correct_ratio']=pd.to_numeric(matches_using_fuzzy_with_Supplier_Number['correct_ratio'], errors='coerce')
matches_using_fuzzy_with_Supplier_Number['correct_ratio'].fillna(0)
matches_using_fuzzy_with_Supplier_Number_above_80=matches_using_fuzzy_with_Supplier_Number[matches_using_fuzzy_with_Supplier_Number['correct_ratio']>80].sort_values(by=['correct_ratio'],ascending=True)
matches_using_fuzzy_with_Supplier_Number_above_80.to_excel('df_6.xlsx')
```
