```python
aud_combo_count_before=len(aud[['Supplier Code','Junior_Buyer']].drop_duplicates())
aud_combo_count_before
```

```python
cols=['Parent_Supplier','Parent_Supplier_name','Supplier_Number','Supplier_Name','Junior_Buyer','Supplier Code','Supplier Name','Site Code','Site Name','Fuzzy_Ratio']
match_2=match[cols][match['Fuzzy_Ratio']!='Tpp Match']
match_2.head(2)
```

```python
final=aud.merge(match_2,how='left',left_on=['Supplier Code','Junior_Buyer'],right_on=['Supplier Code','Junior_Buyer'])
final.head(2)
```

```python
final_matched=final[final['Parent_Supplier'].notnull()]
final_matched.head(2)
```

```python
aud_combo_count=len(final_matched[['Supplier Code','Junior_Buyer']].drop_duplicates())
aud_combo_count
```

```python
print('TQC to COmmercial Match =',(str(round(((aud_combo_count/aud_combo_count_before)*100),2))+"%"))
```
