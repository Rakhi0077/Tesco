```python
import pandas as pd
aud=pd.read_excel(r'C:\Users\IN22912959\OneDrive - Tesco\Documents\Supplier Score Card\New folder\Audits 12 01.xlsx')
spec=pd.read_excel(r'C:\Users\IN22912959\OneDrive - Tesco\Documents\Supplier Score Card\New folder\Old Spec Data (202003).xlsx',sheet_name='Sheet2')
hie=pd.read_excel(r'C:\Users\IN22912959\OneDrive - Tesco\Documents\Supplier Score Card\New folder\hierarchy.xlsx')
```

```python
audits=aud.copy()
spec_tpnb=spec.copy()
hier=hie.copy()
```

```python
audits['Supplier Code']=audits['Supplier Code'].astype(str)
audits['Site Code']=audits['Site Code'].astype(str)
audits['concat']=audits['Supplier Code']+audits['Site Code']
audits.head(2)
# audits.dtypes
```

```python
spec_tpnb['Supplier_Code']=spec_tpnb['Supplier_Code'].astype(str)
spec_tpnb['Site_Code']=spec_tpnb['Site_Code'].astype(str)
spec_tpnb['concat']=spec_tpnb['Supplier_Code']+spec_tpnb['Site_Code']
# spec_tpnb[(spec_tpnb['Site_Code']=='112')&(spec_tpnb['Supplier_Code']=='11937')]
# spec_tpnb.dtypes
# spec_tpnb.head(2)
```

```python
hier.head(2)
```

```python
#getting Spec Tpnb hierarchy based on Tpnb's
spec_hier=spec_tpnb.merge(hier,how='left',left_on=['TPNB'],right_on=['Base_Product_Number'])
spec_hier.head(2)
```

```python
# incorrect tpnb's from spec data hierarchy an its count
incorrect_tpnbs_from_spec=spec_hier[spec_hier['Junior_Buyer'].isnull()]
incorrect_tpnbs_from_spec.head(2)
len(incorrect_tpnbs_from_spec['TPNB'].drop_duplicates())
```

```python
# correct tpnb's from spec data hierarchy and its count
spec_hier_with_correct_tpnbs=spec_hier[spec_hier['Junior_Buyer'].notnull()]
spec_hier_with_correct_tpnbs.head(2)
len(spec_hier_with_correct_tpnbs['TPNB'].drop_duplicates())
```

```python
spec_hier_with_correct_tpnbs.head(2)
cols=['Supplier_Code','Supplier_Name','Site_Code','Site_Name','concat','Junior_Buyer','Buying_Controller','Category_Director','Commercial_Director','Company']
spec_hier_with_correct_tpnbs=spec_hier_with_correct_tpnbs[cols].drop_duplicates()
spec_hier_with_correct_tpnbs.head(2)
```

```python
audits_with_spec_hier=audits.merge(spec_hier_with_correct_tpnbs,how='left',left_on=['concat'],right_on=['concat'])
audits_with_spec_hier.head(2)
```
